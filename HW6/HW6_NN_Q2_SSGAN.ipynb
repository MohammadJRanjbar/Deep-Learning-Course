```python
import torch
from torch import nn
from torch.utils.data import DataLoader, SubsetRandomSampler
from torchvision import datasets, transforms
import torch.optim as optim
from tqdm.auto import tqdm
import matplotlib.pyplot as plt
from torchvision.utils import save_image
import numpy as np
from torchvision import utils
# Hyperparameters
batch_size = 64
learning_rate = 0.0002
epochs = 50
latent_dim = 128
features_gen = 256
features_disc = 128
image_size = 64
channels_img = 1

# Transformations applied on each image
transform = transforms.Compose([
    transforms.Resize(image_size),
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

# Load a subset of the MNIST dataset
subset_size = 10000  # Adjust the size of the subset as needed
full_dataset = datasets.MNIST(root="dataset/", train=True, transform=transform, download=True)
subset_indices = np.random.choice(len(full_dataset), size=subset_size, replace=False)
subset_dataset = torch.utils.data.Subset(full_dataset, subset_indices)
loader = DataLoader(subset_dataset, batch_size=batch_size, shuffle=True, num_workers=4, pin_memory=True)

# Device setting
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Define the Generator and Discriminator models
class Generator(nn.Module):
    def __init__(self, z_dim, channels_img, features_g):
        super(Generator, self).__init__()
        self.gen_size = 8  # Assuming the spatial dimensions start at 8x8
        self.initial = nn.Sequential(
            nn.Linear(z_dim, features_g * self.gen_size * self.gen_size),
            nn.BatchNorm1d(features_g * self.gen_size * self.gen_size),  # Use BatchNorm1d here
            nn.ReLU(),
            nn.Unflatten(1, (features_g, self.gen_size, self.gen_size))
        )
        # rest of the code remains the same...

        self.res_blocks = nn.Sequential(
            ResidualBlock(features_g),
            ResidualBlock(features_g),
            ResidualBlock(features_g)
        )
        self.final = nn.Sequential(
            nn.Conv2d(features_g, channels_img, kernel_size=3, stride=1, padding=1),
            nn.Tanh()
        )

    def forward(self, z):
        x = self.initial(z)
        x = self.res_blocks(x)
        return self.final(x)

# Update the ResidualBlock to match the architecture in the image
class ResidualBlock(nn.Module):
    def __init__(self, in_channels):
        super(ResidualBlock, self).__init__()
        self.block = nn.Sequential(
            nn.BatchNorm2d(in_channels),
            nn.ReLU(),
            nn.Upsample(scale_factor=2, mode='nearest'),
            nn.Conv2d(in_channels, in_channels, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(in_channels),
            nn.ReLU(),
            nn.Conv2d(in_channels, in_channels, kernel_size=3, stride=1, padding=1)
        )
        self.match_dimensions = nn.Sequential(
            nn.Upsample(scale_factor=2, mode='nearest'),
            nn.Conv2d(in_channels, in_channels, kernel_size=1, stride=1, padding=0)
        )

    def forward(self, x):
        x_matched = self.match_dimensions(x)
        return x_matched + self.block(x)


# Residual block for the discriminator
class ResidualDiscBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(ResidualDiscBlock, self).__init__()
        self.block = nn.Sequential(
            nn.Conv2d(in_channels, out_channels, kernel_size=3, stride=2, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.LeakyReLU(0.2),
            nn.Conv2d(out_channels, out_channels, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.LeakyReLU(0.2)
        )

    def forward(self, x):
        return self.block(x)

class Discriminator(nn.Module):
    def __init__(self, channels_img, features_d):
        super(Discriminator, self).__init__()
        self.initial = nn.Sequential(
            ResidualDiscBlock(channels_img, features_d),
            ResidualDiscBlock(features_d, features_d),
            ResidualDiscBlock(features_d, features_d),
            nn.Flatten(),
            nn.Linear(features_d * 8 * 8, 1),  # Adjusted to match the output size
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.initial(x)

def save_fake_images_grid(epoch, generator, latent_dim, num_images=64, format='png'):
    with torch.no_grad():
        # Generate a batch of fake images
        z = torch.randn(num_images, latent_dim).to(device)
        fake_images = generator(z).detach().cpu()

        # Create a grid of images
        grid = utils.make_grid(fake_images, nrow=8, normalize=True)  # nrow is the number of images per row

        # Convert grid to a numpy array
        ndarr = grid.permute(1, 2, 0).numpy()

        # Save the grid as an image file
        plt.imshow(ndarr, interpolation='nearest', cmap='gray')
        plt.axis('off')
        plt.savefig(f'/content/fake_images_epoch_{epoch}.{format}', bbox_inches='tight', pad_inches=0)
        plt.close()
# Initialize Generator and Discriminator models and move them to GPU
netG = Generator(latent_dim, channels_img, features_gen).to(device)
netD = Discriminator(channels_img, features_disc).to(device)

# Initialize weights
def initialize_weights(model):
    for m in model.modules():
        if isinstance(m, (nn.Conv2d, nn.ConvTranspose2d, nn.Linear)):
            nn.init.normal_(m.weight.data, 0.0, 0.02)

# Setup Adam optimizers for both G and D
optimizerD = optim.Adam(netD.parameters(), lr=learning_rate, betas=(0.5, 0.999))
optimizerG = optim.Adam(netG.parameters(), lr=learning_rate, betas=(0.5, 0.999))

# Binary Cross Entropy loss function
criterion = nn.BCELoss()


# Training Loop
g_losses = []
c_losses = []
accumulation_steps = 4  # Accumulate gradients over 4 mini-batches

for epoch in range(epochs):
    running_lossG = 0.0
    running_lossD = 0.0

    for batch_idx, (real, _) in enumerate(tqdm(loader)):
        real = real.to(device)
        batch_size = real.size(0)
        noise = torch.randn(batch_size, latent_dim).to(device)

        netD.zero_grad()
        label = (torch.ones(batch_size) * 0.9).to(device)
        output = netD(real).reshape(-1)
        lossD_real = criterion(output, label)
        D_x = output.mean().item()

        fake = netG(noise)
        label = (torch.zeros(batch_size) * 0.1).to(device)
        output = netD(fake.detach()).reshape(-1)
        lossD_fake = criterion(output, label)

        lossD = lossD_real + lossD_fake
        lossD.backward()

        if (batch_idx + 1) % accumulation_steps == 0:
            optimizerD.step()
            netD.zero_grad()
            netG.zero_grad()

            label = torch.ones(batch_size, device=device)
            output = netD(fake).reshape(-1)
            lossG = criterion(output, label)
            lossG.backward()
            optimizerG.step()
        else:
            optimizerD.step()
            lossG = 0  # Initialize lossG

        running_lossG += lossG.item() if lossG != 0 else 0
        running_lossD += lossD.item()

    # Calculate average loss for the epoch
    avg_lossG = running_lossG / len(loader)
    avg_lossD = running_lossD / len(loader)

    # Save generated images after each epoch
    with torch.no_grad():
        save_fake_images_grid(epoch + 1, netG, latent_dim)

    g_losses.append(avg_lossG)
    c_losses.append(avg_lossD)

    # Print progress
    print(f'Epoch [{epoch + 1}/{epochs}] - Loss C: {avg_lossD:.4f}, Loss G: {avg_lossG:.4f}')

# Plot the losses
plt.figure(figsize=(10,5))
plt.title("Generator and Discriminator Loss During Training")
plt.plot(g_losses, label="G")
plt.plot(c_losses, label="D")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()

# Save the final batch of fake images
with torch.no_grad():
    save_fake_images_grid('final', netG, latent_dim)

print("Training is complete!")
```

```python
import os
import torch
import torchvision
from torch.utils.data import DataLoader
from pytorch_fid import fid_score


path_gen_images="generated_images"
path_real_images="real_images"
# Ensure the directories exist
os.makedirs(path_gen_images, exist_ok=True)
os.makedirs(path_real_images, exist_ok=True)

# Generate images using the trained generator model
with torch.no_grad():
    latent_vectors = torch.randn(128, latent_dim).to(device)
    generated_imgs = netG(latent_vectors)

# Save generated images
for idx, gen_img in enumerate(generated_imgs):
    torchvision.utils.save_image(gen_img, os.path.join(path_gen_images, f'gen_img_{idx + 1}.png'))

# Collect and save a subset of real images from the dataset
saved_real_images = []
target_num_images = 128

for batch_labels, batch_images in loader:
    saved_real_images.extend(batch_labels)
    if len(saved_real_images) >= target_num_images:
        break

for idx, image in enumerate(saved_real_images[:target_num_images]):
    torchvision.utils.save_image(image, os.path.join(path_real_images, f'real_img_{idx + 1}.png'))

# FID Score Calculation
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
batch_size_fid = 32
fid_score_value = fid_score.calculate_fid_given_paths([path_real_images, path_gen_images], batch_size=batch_size_fid, device=device, dims=2048)
print(f"FID Score: {fid_score_value}")
```
