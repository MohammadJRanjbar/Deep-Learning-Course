```python

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score
from sklearn.utils import shuffle

import hazm
#from cleantext import clean

import plotly.express as px
import plotly.graph_objects as go

from tqdm.notebook import tqdm

import os
import re
import json
import copy
import collections
import re
from hazm import Normalizer, word_tokenize,Stemmer
from hazm import stopwords_list, word_tokenize
from nltk.stem import PorterStemmer
import matplotlib.pyplot as plt
import os
from sklearn.metrics import confusion_matrix
import torch
import torch.nn as nn
from transformers import AutoConfig, AutoTokenizer, AutoModel
import random
```

```python
def preprocess(data):
  normalizer = Normalizer()
  stemmer = PorterStemmer()
  for i in range(len(data)):
    text=data[i]
    tokens = word_tokenize(text)
    filtered_tokens = [token for token in tokens if token not in stopwords_list()]
    filtered_text = ' '.join(filtered_tokens)
    filtered_text = filtered_text.replace("\u200c", "")
    data[i]=normalizer.normalize(filtered_text)
  return data
```

```python
train_df["premise"]=preprocess(train_df["premise"])
train_df["hypothesis"]=preprocess(train_df["hypothesis"])

test_df["premise"]=preprocess(test_df["premise"])
test_df["hypothesis"]=preprocess(test_df["hypothesis"])

val_df["premise"]=preprocess(val_df["premise"])
val_df["hypothesis"]=preprocess(val_df["hypothesis"])
```

```python
from datasets import Dataset
train_dataset = Dataset.from_pandas(train_df)
valid_dataset = Dataset.from_pandas(val_df)
test_dataset  = Dataset.from_pandas(test_df)
```

```python
import torch
import pandas as pd
import numpy as np
from transformers import AutoTokenizer, AutoModelForSequenceClassification, AdamW
from torch.utils.data import DataLoader, Dataset
from sklearn.model_selection import train_test_split

def tokenizer_function(train_df):
    max_length = 128  # Set your desired max length
    tokenizer = AutoTokenizer.from_pretrained("HooshvareLab/bert-base-parsbert-uncased", max_length=max_length)
    tokenized_batch = tokenizer(train_df["premise"], train_df["hypothesis"], padding='max_length', max_length=max_length,  truncation=True, return_tensors='pt')
    return tokenized_batch
train_dataset = train_dataset.map(tokenizer_function, batched=True)
valid_dataset = valid_dataset.map(tokenizer_function, batched=True)
test_dataset  = test_dataset .map(tokenizer_function, batched=True)
```

```python
train_dataset.set_format('torch', columns=["input_ids", "attention_mask", "label"] )
test_dataset.set_format('torch', columns=["input_ids", "attention_mask", "label"] )
valid_dataset.set_format('torch', columns=["input_ids", "attention_mask", "label"] )
```

```python
train_loader = DataLoader(train_dataset, shuffle = True, batch_size = 32)
val_loader   = DataLoader(valid_dataset, shuffle = True, batch_size = 32)
test_loader  = DataLoader(test_dataset, shuffle = True, batch_size = 32)
```

# Fine Tuning

```python
from transformers import AutoModelForSequenceClassification, AutoConfig

model = AutoModelForSequenceClassification.from_pretrained("HooshvareLab/bert-base-parsbert-uncased", num_labels=3)
```

```python
print(model)
```

```python
from tqdm import tqdm
import torch
from torch.optim import AdamW
from torch.utils.data import DataLoader

# Move model to device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Fine-tuning parameters
optimizer = AdamW(model.parameters(), lr=2e-5)
epochs = 5

# Define a mapping from label strings to numerical values
label_mapping = {"n": 0, "e": 1, "c": 2}  # Adjust according to your label classes

# Fine-tuning loop
# Lists to store training and validation metrics
train_loss_list = []
train_accuracy_list = []
val_loss_list = []
val_accuracy_list = []

# Fine-tuning loop
for epoch in range(epochs):
    model.train()
    train_loss = 0
    correct_predictions = 0
    total_samples = 0

    # Use tqdm to create a progress bar for training
    train_loader_tqdm = tqdm(train_loader, desc=f'Epoch {epoch + 1}/{epochs} - Training', leave=False)

    for batch in train_loader_tqdm:
        optimizer.zero_grad()
        input_ids, attention_mask, labels = batch['input_ids'], batch['attention_mask'], batch['label']

        # Convert labels to numerical values
        labels = [label_mapping[label_str] for label_str in labels]

        # Convert to PyTorch tensors and move to device
        input_ids, attention_mask, labels = (
            torch.tensor(input_ids).to(device),
            torch.tensor(attention_mask).to(device),
            torch.tensor(labels).to(device)
        )

        outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs['loss']  # Accessing loss through dictionary key
        train_loss += loss.item()
        loss.backward()
        optimizer.step()

        # Calculate accuracy
        _, predicted_labels = torch.max(outputs.logits, 1)
        correct_predictions += (predicted_labels == labels).sum().item()
        total_samples += labels.size(0)

        # Update tqdm progress bar
        train_loader_tqdm.set_postfix(loss=train_loss / total_samples, accuracy=correct_predictions / total_samples)

    # Calculate training accuracy and loss
    avg_train_loss = train_loss / total_samples
    train_accuracy = correct_predictions / total_samples

    # Validation
    model.eval()
    val_loss = 0
    correct_predictions_val = 0
    total_samples_val = 0

    with torch.no_grad():
        for batch in val_loader:
            input_ids, attention_mask, labels = batch['input_ids'], batch['attention_mask'], batch['label']

            # Convert labels to numerical values
            labels = [label_mapping[label_str] for label_str in labels]

            # Convert to PyTorch tensors and move to device
            input_ids, attention_mask, labels = (
                torch.tensor(input_ids).to(device),
                torch.tensor(attention_mask).to(device),
                torch.tensor(labels).to(device)
            )

            outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
            val_loss += outputs['loss'].item()  # Accessing loss through dictionary key

            # Calculate accuracy for validation
            _, predicted_labels_val = torch.max(outputs.logits, 1)
            correct_predictions_val += (predicted_labels_val == labels).sum().item()
            total_samples_val += labels.size(0)

    avg_val_loss = val_loss / total_samples_val
    val_accuracy = correct_predictions_val / total_samples_val

    # Append metrics to lists
    train_loss_list.append(avg_train_loss)
    train_accuracy_list.append(train_accuracy)
    val_loss_list.append(avg_val_loss)
    val_accuracy_list.append(val_accuracy)

    print(f'Epoch {epoch + 1}/{epochs} - Validation Loss: {avg_val_loss}, Validation Accuracy: {val_accuracy}')

# Save the fine-tuned model
model.save_pretrained("/path/to/save/fine-tuned-model")
```

```python

# Plotting
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(train_loss_list, label='Training Loss')
plt.plot(val_loss_list, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(train_accuracy_list, label='Training Accuracy')
plt.plot(val_accuracy_list, label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```

```python
# Evaluate on test set
from sklearn.metrics import f1_score

model.eval()
test_labels = []
predicted_labels_list = []

with torch.no_grad():
    for batch in test_loader:
        input_ids, attention_mask, labels = batch['input_ids'], batch['attention_mask'], batch['label']

        # Convert labels to numerical values
        labels = [label_mapping[label_str] for label_str in labels]

        # Convert to PyTorch tensors and move to device
        input_ids, attention_mask, labels = (
            torch.tensor(input_ids).to(device),
            torch.tensor(attention_mask).to(device),
            torch.tensor(labels).to(device)
        )

        outputs = model(input_ids, attention_mask=attention_mask, labels=labels)

        # Collect ground truth labels and predicted labels
        test_labels.extend(labels.cpu().numpy())
        predicted_labels_list.extend(torch.argmax(outputs.logits, 1).cpu().numpy())

# Calculate confusion matrix and classification report
conf_matrix = confusion_matrix(test_labels, predicted_labels_list)
class_report = classification_report(test_labels, predicted_labels_list)

import seaborn as sns
import matplotlib.pyplot as plt

# Plot confusion matrix
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=label_mapping.keys(), yticklabels=label_mapping.keys())
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()


# Print and display classification report
print("Classification Report:")
print(class_report)

f1 = f1_score(test_labels, predicted_labels_list, average='weighted')

# Print and display F1 score
print("F1 Score:", f1)
```
