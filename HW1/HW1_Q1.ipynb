```python
import numpy as np
import itertools
class McCulloch_Pitts_neuron():

  def __init__(self , weights , threshold):
    self.weights = weights    #define weights
    self.threshold = threshold    #define threshold

  def model(self , x):
    #define model with threshold
    if np.dot(self.weights, x) >= self.threshold:
      return 1
    else:
      return 0
num_cols = 7
num_rows = 128

arr = np.zeros((num_rows, num_cols), dtype=int)

for i in range(2**num_cols):
  row = np.binary_repr(i, width=num_cols)
  row = [int(x) for x in row]

  arr[i] = row

X= np.array(arr)
print(X.shape , X.dtype, X )
```

```python
n6 = McCulloch_Pitts_neuron([1,-1,1,1,1,1,1], 6)
n7 = McCulloch_Pitts_neuron([1,1,1,-1,-1,-1,-1], 3)
n8 = McCulloch_Pitts_neuron([1,1,1,1,1,1,1], 7)
n9 = McCulloch_Pitts_neuron([1,1,1,1,-1,1,1], 6)
for inp in X:

  o1 = n6.model(inp)
  o2 = n7.model(inp)
  o3 = n8.model(inp)
  o4 = n9.model(inp)

  if o1:
    print(inp, "- Digit 6")
  elif o2:
    print(inp, "- Digit 7")
  elif o3:
    print(inp, "- Digit 8")
  elif o4:
    print(inp, "- Digit 9")
  else:
    print(inp, "- No digit detected")
```
